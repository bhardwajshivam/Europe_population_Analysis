```python
def death_rate(s,e,t):
    dr = (s-e)/t
    return round(dr, 2)
```

# 7.

```python
table = []
for i in range(1, len(data)):
    d = data[i].split(',')
    for j in range(1, len(d)):
        d[j] = int(d[j])
    table.append(d[1:])
```

```python
table
```

```python
summary10 = []
for col in range(len(table[0])):
    x = []
    for row in range(10,len(table),10):
        x.append(death_rate(table[row-10][col],table[row][col],10))
    summary10.append(x)
```

```python
summary10
```

# 8.

```python
rank_matrix = []
for j in range(len(summary10[0])):
    colj = []
    for k in range(len(summary10)):
        colj.append(summary10[k][j])  
    rj = [i for i, x in sorted(enumerate(colj), key=lambda x: x[1])]
    rank_matrix.append(rj)
```

```python
rank_matrix
```

```python
oneToTen = rank_matrix[0]
```

```python
l3_oneToTen = []
for i in range(3):
    l3_oneToTen.append(rc[oneToTen[i]])
```

# 10. 

```python
count = []
for j in range(len(rank_matrix[0])):
    c = 0
    for i in range(1,len(rank_matrix)):
        if rank_matrix[i][j] != rank_matrix[i-1][j]:
            c+=1
    count.append(c)
```

```python
count
```

```python
most_changes = []
for i in range(len(count)):
    if count[i] == max(count):
        most_changes.append(rc[i])
```

```python
most_changes
```
